```python
import itertools
import os
import shutil

import matplotlib.pylab as plt
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

database_dir = "drive/MyDrive/Sodankyla_database-20211027T162911Z-002/Sodankyla_database";
```

```python
                    
good_days_src_file_dir = database_dir + "/aurora_list.txt";                     # File to look through to find the good days
good_days_new_dir      = database_dir + "/DP_good_days"                         # Dir for collecting all the good images in a single batch
long_log = False;                                                               # Log all of the files and results found



    # Extract all files in the list that have "good day" in them somewhere
good_days = [];
if long_log: print("'Good' days: ")
with open(good_days_src_file_dir, 'r') as days:
  contents = days.read().split('\n')
 
  for lineno in range(len(contents)):
    line = contents[lineno]
    if "good day" in line:
      line_date = line.split('-');
      if long_log: print("   -" + line_date[1][1:-1]);                          # We have found and extracted the correct part of the filename # 
      good_days.append(line_date[1][1:-1]);


all_good_day_images = []
    # Look at all the folders and get all images that were on 'good' days
images = os.listdir(database_dir + "/RAW/")
for day_ext in good_days:
  year = "20"+day_ext[-2:]
  image_list = os.listdir(database_dir + "/RAW/" + year + '/')

  if long_log: print("Images from " + day_ext + ':')
  for image_dir in image_list:
    #print(day_ext[:-1] + "     vs     " + image_dir + "     gives: " + str(str(day_ext[:-1]) in str(image_dir)))
    if day_ext[:-1] in image_dir:
      if long_log: print("  " + image_dir);
      all_good_day_images.append(database_dir + "/RAW/" + year + '/' + image_dir)  # We have found the images with dates in the filenames that are on the list of good days #
  if long_log: print("\n\n")


print("Found " + str(len(all_good_day_images)) + " good images")
print("Found " + str(len(os.listdir(good_days_new_dir))) + " good images already in the batch")

print("Copying images over to " + good_days_new_dir)                            # We now check the folder for the files we are...
                                                                                # ...going to add and only add if they arent already in there
existing_images = os.listdir(good_days_new_dir)
for image_dir in all_good_day_images:
  if long_log: print(image_dir)
  if image_dir.split('/')[6] not in existing_images:
    shutil.copy2(image_dir,good_days_new_dir)
    print(image_dir.split('/')[6] + " is not already in the batch")

```
