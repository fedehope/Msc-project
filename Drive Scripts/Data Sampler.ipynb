#Data Sampler

##Purpose
<p>The purpose of this script is to automate the process of selecting images from RAW datasets to be put into folders. This is initially being done to make a dataset for use by the crop and cutout tool to make new types of training data</p>

##Structure
###1: Set variables for key dirs
###2: Search through those dirs for data and copy to the new dir


#Import libraries

```python
from google.colab import drive
drive.mount('/content/drive',force_remount=True)

import numpy as np
import os
import shutil

import random as rd
```

#Variable definition (locations, amounts etc.)

```python
# Location of the raw data to sample from
raw_dir   = "/content/drive/MyDrive/Sodankyla_database-20211027T162911Z-002/Sodankyla_database/RAW"
# Location to put the selected images for processing
final_dir = "/content/drive/MyDrive/Sodankyla_database-20211027T162911Z-002/Sodankyla_database/Testing Auto-Cropping and Cutouts/Executable crop tool/application.windows64/data/uncropped_images"

# Total amount to sample across all available data
total_sample_number = 1000
```

# Selection of images from raw 

```python
total_copied = 0

total_to_be_sampled_number = total_sample_number - len(os.listdir(final_dir))

# Iterate through all the years and see what the max possible sampling is
year_list = os.listdir(raw_dir)
print(year_list)
sizes = {}
ins_years = []
for year in year_list:
  spec_year_dir = raw_dir+'/'+year
  contents = os.listdir(spec_year_dir)
  sizes[year] = len(contents)
  if sizes[year] < total_to_be_sampled_number/len(year_list):
    ins_years.append(year)

# Ignore all years from the ins_years list (for now)

# Iterate through the years in the RAW folder and select some images from each 
# based on randomly sampling the size of the list
for year in year_list:
  if year not in ins_years:
    spec_year_dir = raw_dir+'/'+year
    contents = os.listdir(spec_year_dir)
    n_images = len(contents)
    for i in range(   int(total_to_be_sampled_number/( len(year_list)-len(ins_years) ))   ):
      selection   = rd.randint(0,n_images-1)
      source      = spec_year_dir+'/'+contents[selection]
      destination = final_dir
      #print("Source: "+source)
      #print("Destination: "+destination)
      if contents[selection] in os.listdir(final_dir):
        print(source + "not copied (already in folder!)")
      else:
        shutil.copy(source,destination)
  print(year)

print(ins_years)
print(os.listdir(final_dir))
  
```
