```python
import numpy as np
import os
from sklearn.model_selection import train_test_split
import shutil

from google.colab import drive

drive.mount('/content/drive', force_remount=True)
```

```python
#############################################
# Iterates over each class folder, splits the images in the folder into train a test set 80:20
# and copies them into train and test set folders
#############################################


classes = ['Arcs',
           'Breakup',
           'Colored',
           'Discrete',
           'Edge',
           'Faint',
           'Patchy']


folder = '/content/drive/MyDrive/Sodankyla_database-20211027T162911Z-002/Sodankyla_database/DP_good_days/labelled'

# iterating over each class folder
for clas in classes:
    clas_dir = os.path.join(folder, 'classes', clas)
    img_files = os.listdir(clas_dir)

    # removing the desktop.ini file if it exists 
    if os.path.exists(os.path.join(clas_dir, 'desktop.ini')):
        img_files.remove('desktop.ini')

    train, test =  train_test_split(img_files, test_size = 0.2, random_state = 42) # splits the set
    print(train)
    # coping training images
    for img_file in train:
        source = os.path.join(clas_dir, img_file)
        destination = os.path.join(folder, 'train_set', img_file)
        print(source)
        print(destination)
        # copying image
        shutil.copy(source, destination)

    # copying test images
    for img_file in test:
        source = os.path.join(clas_dir, img_file)
        destination = os.path.join(folder, 'test_set', img_file)

        # copying image
        shutil.copy(source,destination)
```
